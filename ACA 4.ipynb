```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Visualizing the network

We have tracked each classmate’s responses and posts within the class discussion to see who commented to whom. This discussion board does not require the original poster to reply to the commenters, so it is a one way relationship, and we may see one poster may get the most comments from different classmates. 


```python
SNAdata = pd.read_csv("ACA4_data.csv", index_col = 0)
G = nx.Graph(SNAdata)
```

```python
nx.draw(G, with_labels = True)
```

# Network Statistics


# Density

```python
density = nx.density(G)

print('The edge density is: %.3f' % (density)) ### Here I am using a format for the output decimal places
                                             ### See tutorial here: https://www.geeksforgeeks.org/python-output-formatting/
```

This value of 0.143 indicates that this network is not well connected because we can see less than 20% links among all possible links. This is probably because the class does not require orginal poster to reply to the commenters. This is a one way commenting system, but we can see there are people who has most links, may have a greater centrality within this network than others. Examples as Ty Scaletta and Xiaoyue Zhang 

# Degree Centrality

```python
degree = nx.degree(G)
degree
```

```python
pd.DataFrame(degree)
```

Ty Scaletta	tend to have highest degree of centrality, most people commeted on her post and know her.

# Closeness centrality

```python
close = nx.closeness_centrality(G)
pd.DataFrame.from_dict(close, orient = 'index')
```

The higher value of closenss means the higher centrality, in this case Ty Scaletta is also the central. 

# Betweenness centrality

```python
between = nx.betweenness_centrality(G)
pd.DataFrame.from_dict(between, orient = 'index')
```

Another way to measure the centrality, Ty Scaletta is proved to be the center. 

# Community detection

```python
from networkx.algorithms.community.modularity_max import greedy_modularity_communities

c = list(greedy_modularity_communities(G))

print(len(c))
```

there is 5 communities

```python
#Lets see these 3 clusters
community_0 = sorted(c[0])
community_1 = sorted(c[1])
community_2 = sorted(c[2])
community_3 = sorted(c[3])
community_4 = sorted(c[4])


print(community_0)
print(community_1)
print(community_2)
print(community_3)
print(community_4)
```

```python
# Set seed (optional)
np.random.seed(12345)

# Obtain the positions of the nodes
pos = nx.spring_layout(G)

# Print out all the nodes with color
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_0, node_color='green', label = True)
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_1, node_color='red', label = True)
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_2, node_color='blue', label = True)
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_3, node_color='pink', label = True)
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_4, node_color='purple', label = True)
# Print out the edges
nx.draw_networkx_edges(G, pos = pos, width = 0.5)

# Make sure the label is printed
nx.draw_networkx_labels(G, pos=pos)

# Show the plot so it supress the other networkx outputs
plt.show()
```

```python
nx.draw_networkx_nodes(G, pos = pos, nodelist = dict(degree).keys(), node_size = [v * 50 for v in dict(degree).values()], label = True)
nx.draw_networkx_edges(G, pos = pos, width = 0.2)
nx.draw_networkx_labels(G, pos=pos)
plt.show()
```

```python
eWeight = nx.get_edge_attributes(G,'weight')

nx.draw_networkx_nodes(G, pos = pos, nodelist = dict(degree).keys(), node_size = [v * 50 for v in dict(degree).values()], label = True)
nx.draw_networkx_edges(G,pos, width = [5**e/10 for e in eWeight.values()])
nx.draw_networkx_labels(G, pos=pos)

plt.show()
```

```python
# Degrees of each community
community_0_degree = {c: dict(degree)[c] for c in community_0}
community_1_degree = {c: dict(degree)[c] for c in community_1}
community_2_degree = {c: dict(degree)[c] for c in community_2}
community_3_degree = {c: dict(degree)[c] for c in community_3}
community_4_degree = {c: dict(degree)[c] for c in community_4}


# Print out the nodes with color and size
nx.draw_networkx_nodes(G, pos = pos, 
                       nodelist=community_0,
                       node_color='green',
                       node_size = [v * 50 for v in community_0_degree.values()],
                       label = True, node_shape="s", alpha=0.5)

nx.draw_networkx_nodes(G, pos =pos,
                       nodelist=community_1,
                       node_size = [v * 50 for v in community_1_degree.values()],
                       node_color='red', label = True, node_shape="s", alpha=0.5)

nx.draw_networkx_nodes(G, pos = pos,
                       nodelist=community_2,
                       node_size = [v * 50 for v in community_2_degree.values()],
                       node_color='lightblue', label = True,  node_shape="s", alpha=0.5)

nx.draw_networkx_nodes(G, pos = pos,
                       nodelist=community_3,
                       node_size = [v * 50 for v in community_3_degree.values()],
                       node_color='pink', label = True,  node_shape="s", alpha=0.5)

nx.draw_networkx_nodes(G, pos = pos,
                       nodelist=community_4,
                       node_size = [v * 50 for v in community_4_degree.values()],
                       node_color='purple', label = True,  node_shape="s", alpha=0.5)

# Print out the edges with proper width
nx.draw_networkx_edges(G,pos, width = [10**e/10 for e in eWeight.values()])

# Make sure the label is printed
nx.draw_networkx_labels(G, pos=pos)

# Show the plot so it supress the other networkx outputs
plt.show()
```
