# # Selecting variables 


```python
score_card = Scorecard.dropna(subset=['ADM_RATE','ADM_RATE_ALL'])
```

```python
score_card1 = score_card.dropna(axis= "columns")
score_card1
```

```python
score_card2 = score_card1.drop(score_card1.columns[score_card1.columns.str.match(r'PrivacySupressed')], axis=1)
```

```python
score_card2
```

```python
from scipy import stats
%matplotlib notebook
```

```python
score_card3 = score_card2.drop(['INSTNM','CITY','STABBR','ZIP'], axis=1)
score_card3
```

```python
New_score_card= np.log(score_card3.iloc[:,1:]) 
New_score_card
```

```python
clean_score_card = New_score_card.replace([np.inf, -np.inf], np.nan)
clean_score_card
```

```python
limitPer = len(clean_score_card) * .8
clean_score_card1= clean_score_card.dropna(thresh=limitPer, axis=1)
clean_score_card1
```

```python
clean_score_card2 = clean_score_card1.drop('main',1)
clean_score_card3 = clean_score_card2.dropna(axis= "rows")
```

```python
clean_score_card3
```

# Scree Plot Evaluation:
    

```python
# Import the standard stuff
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(clean_score_card3)
```

```python
pca = PCA()

pca.fit(clean_score_card3)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize = (10,2))
plt.plot(range(1,19), pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('scree plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
print(plt.show())
```

```python
plt.figure(figsize = (10,2))
plt.plot(range(1,19), pca.explained_variance_ratio_.cumsum(), 'o-', linewidth=2, color='blue')
plt.title('Variance explained by the components')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
print(plt.show())
```

We can see that there are 6 components needed to preserve 80 percent of the variance. 

As the PCA result of ACA 3, I have used 8 components at the begining and down to 3 components.
